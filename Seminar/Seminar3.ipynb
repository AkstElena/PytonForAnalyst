```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', sep=',', encoding='Windows-1251')
df.head()
```

1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu_Company']=df['Cpu'].apply(lambda x: x.split()[0])
df.head()
```

1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def only_gb(x):
    listNew = x.split()
    if 'TB' in listNew[0]:
        result = listNew[0].replace('TB', '')
        result = float(result)*1024
    elif 'GB' in listNew[0]:
        result = float(listNew[0].replace('GB', ''))
    return result
```

```python
df['Memory_Amount']=df['Memory'].apply(only_gb)
df.sample(5)
```

1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
def only_mem_type(x):
    list_split = x.split()
    result_ = []
    for i in list_split:
        if "GB" not in i and "TB" not in i:
            result_.append(i)
    result = " ".join(result_)
    return result
```

```python
df['Memory_Type'] = df['Memory'].apply(only_mem_type)
df.sample(5)
```

1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.head()
```

2.2 Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD

```python
df['SSD']=0
df.head()
```

```python
df.loc[df['Memory_Type']=='SSD','SSD']=1
df.sample(5)
```

```python
df.loc[df['Memory_Type']=='SSD + HDD','SSD']=1
df.sample(5)
```

2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный

```python
df['Weight'] = df['Weight'].str.replace('kg','')
df['Weight'] = df['Weight'].astype(float)
```
