```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1.1. Скачать следующие данные: kc-house-data

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df.info()
```

1.2. Изучите стоимости недвижимости

```python
plt.hist(df['price'])
plt.title('Стоимость недвижимости на побережье Калифорнии')
plt.xlabel('Цена недвижимости')
plt.ylabel('Количество объектов');
```

В основном вся недвижимость находится в диапазоне цены до 1 млн.

1.3. Изучите распределение квадратуры жилой

```python
data = df['sqft_living'].value_counts()
data
```

```python
plt.scatter(data.index, data.values)
plt.title('Распределение жилой площади')
plt.xlabel('Размер жилой площади')
plt.ylabel('Количество объектов');
```

Наибольшее количество домов с жилой площадью в районе от 1000 до 3000

1.4. Изучите распределение года постройки

```python
data = df['yr_built'].value_counts()
data
```

```python
plt.bar(data.index, data.values)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов');
```

Преобладают дома построенные после 1950ых годов, так же значительное количество домов построено после 2010ого года

```python
data = df['waterfront'].value_counts()
data
```

2.2. Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
data
```

```python
plt.bar(data.index, data.values)
plt.title('Распределение объектов недвижимости в зависимости от этажности')
plt.xlabel('Количество этажей')
plt.ylabel('Количество объектов');
```

2.3. Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
data
```

```python
plt.pie(data.values, autopct='%1.2f%%', explode=(0.1, 0.1, 0.1,0.3,0.5), startangle=90 )
plt.legend(data.index)
plt.axis('equal')
plt.title('Состояние домов');
```

Почти 65% домов оцениваются на 3, только 0,14% в очень плохом состоянии. Наивысшая оценка у 7,87% домов.

## Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.\
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')

plt.xlabel('Цена')
plt.ylabel('Количество ванных')
plt.title('Зависимость цены от количества ванных');
```

Наблюдается линейная зависимость цены от количества ванных комнат

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');

plt.xlabel('Широта')
plt.ylabel('Долгота')
plt.title('Соотношение широты и долготы');
```

Большинство домов соредоточено между 47,5 и 47,8 градусами широты. 

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg')

plt.xlabel('Цена')
plt.ylabel('Количество этажей')
plt.title('Зависимость цены от количества этажей');
```

Наблюдается линейная зависимость цены от количества этажей

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')

plt.xlabel('Цена')
plt.ylabel('Количество спален')
plt.title('Зависимость цены от количества спален');
```

Наблюдается линейная зависимость цены от количества спален

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')

plt.xlabel('Цена')
plt.ylabel('Оценка')
plt.title('Зависимость цены от состояния');
```

Наблюдается линейная зависимость цены от состояния
